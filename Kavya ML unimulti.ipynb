```python
#Univariate Analysis for Categorical Variables
#1 Accident Severity Distribution:
import seaborn as sns
import matplotlib.pyplot as plt

# Count plot for Accident_Severity
sns.countplot(data=df, x='Accident_Severity')
plt.title('Accident Severity Distribution')
plt.show()
```

Univariate Analysis for Numerical Variables

```python
#Univariate Analysis for Numerical Variables
#Summary Statistics (Mean, Median, Mode, etc.):
# Example: For Speed_limit column
speed_limit = df['Speed_limit']
mean_speed = speed_limit.mean()
median_speed = speed_limit.median()
mode_speed = speed_limit.mode()[0]  # Mode might return a series, so take the first value
std_speed = speed_limit.std()
iqr_speed = speed_limit.quantile(0.75) - speed_limit.quantile(0.25)

# Display summary statistics
print(f"Speed Limit Summary:\nMean: {mean_speed}\nMedian: {median_speed}\nMode: {mode_speed}")
print(f"Standard Deviation: {std_speed}\nIQR: {iqr_speed}")
```

```python
# Histogram
sns.histplot(df['Speed_limit'], bins=10, kde=True)
plt.title('Speed Limit Distribution')
plt.show()

# Box plot
sns.boxplot(data=df, x='Speed_limit')
plt.title('Box Plot of Speed Limit')
plt.show()
```

```python
# Five Number Summary (Min, Q1, Median, Q3, Max) for Speed Limit
five_num_summary = df['Speed_limit'].describe()
print(f"Five Number Summary:\n{five_num_summary}")
```

```python
mean = df['Speed_limit'].mean()
std_dev = df['Speed_limit'].std()

# 68% Confidence Interval (Mean ± 1 Std Dev)
conf_68 = (mean - std_dev, mean + std_dev)

# 95% Confidence Interval (Mean ± 2 Std Dev)
conf_95 = (mean - 2 * std_dev, mean + 2 * std_dev)

print(f"68% of data lies between: {conf_68}")
print(f"95% of data lies between: {conf_95}")
```

```python
sns.boxplot(data=df, x='Speed_limit')
plt.title('Box Plot for Speed Limit')
plt.show()
```

```python
#Box Plot to Identify Outliers
sns.boxplot(data=df, x='Accident_Severity', y='Speed_limit')
plt.title('Accident Severity vs Speed Limit')
plt.show()
```

```python
#Categorical vs Numerical Variable Analysis:
sns.boxplot(data=df, x='Accident_Severity', y='Speed_limit')
plt.title('Accident Severity vs Speed Limit')
plt.show()
```

```python
# Libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv('/Accident_Information.csv')

# Display the first few rows
print(df.head())

# Multivariate Analysis

# Correlation Matrix for Numerical Variables
numerical_columns = ['Number_of_Casualties', 'Number_of_Vehicles', 'Speed_limit', 'Latitude', 'Longitude']
correlation_matrix = df[numerical_columns].corr()

# Heatmap of Correlation Matrix
plt.figure(figsize=(10, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Matrix of Numerical Variables')
plt.show()

# Pair Plot for selected numerical variables
sns.pairplot(df[numerical_columns])
plt.title('Pair Plot of Numerical Variables')
plt.show()

# Crosstab for Categorical Variables
severity_vs_weather = pd.crosstab(df['Accident_Severity'], df['Weather_Conditions'])
print(severity_vs_weather)

# Boxplots of Numerical Variables Grouped by Accident Severity
plt.figure(figsize=(10, 6))
sns.boxplot(x='Accident_Severity', y='Number_of_Casualties', data=df)
plt.title('Boxplot of Number of Casualties by Accident Severity')
plt.show()

# Summary Statistics
print(df[numerical_columns].describe())

# Standard Deviation and IQR
std_dev = df[numerical_columns].std()
iqr = df[numerical_columns].quantile(0.75) - df[numerical_columns].quantile(0.25)

print("\nStandard Deviation:\n", std_dev)
print("\nInterquartile Range (IQR):\n", iqr)

# Rule of 68-95% for Number of Casualties
mean_casualties = df['Number_of_Casualties'].mean()
std_casualties = df['Number_of_Casualties'].std()

rule_68 = (mean_casualties - std_casualties, mean_casualties + std_casualties)
rule_95 = (mean_casualties - 2*std_casualties, mean_casualties + 2*std_casualties)

print(f"\n68% range for Number of Casualties: {rule_68}")
print(f"95% range for Number of Casualties: {rule_95}")

# Boxplot for Speed Limit by Severity
plt.figure(figsize=(10, 6))
sns.boxplot(x='Accident_Severity', y='Speed_limit', data=df)
plt.title('Boxplot of Speed Limit by Accident Severity')
plt.show()
```
